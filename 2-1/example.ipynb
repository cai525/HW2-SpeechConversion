## 参数配置

### 音频参数

```python
src_path = './archive/data/wav_test/p1/p1_331.wav'  # 音频目录
speakers = ['p1', 'p2']  # 语者
speakers_id = {speaker: i for i, speaker in enumerate(speakers)}  # 语者id:应与训练所用id对应
fs = 16000  # 采样率
```
